```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, PowerTransformer
```

```python
df = pd.read_csv("academic_performance.csv")

print("Original Dataset:")
print(df)
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
df["Maths"] = df["Maths"].fillna(df["Maths"].mean())
df["Science"] = df["Science"].fillna(df["Science"].mean())
df["English"] = df["English"].fillna(df["English"].mean())
df["Attendance"] = df["Attendance"].fillna(df["Attendance"].mean())
```

```python
df["Attendance"] = df["Attendance"].clip(0, 100)
```

```python
print("dataset After Handling Missing Values:")
print(df)
```

```python
numeric_cols = ["Maths", "Science", "English", "Attendance"]

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)  #A quantile divides sorted data into equal parts.
    Q3 = df[col].quantile(0.75   #It tells you the value below which a certain percentage of data falls.
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"\nOutliers in {col}:")
    print(outliers[[col]])

    # Replace outliers with median
    median_value = df[col].median()
    df.loc[(df[col] < lower) | (df[col] > upper), col] = median_value

print("\nDataset After Outlier Treatment:")
print(df)
```

```python
scaler = MinMaxScaler()  #MinMaxScaler is used to rescale numerical data into a fixed range, usually 0 to 1.
df[["Attendance"]] = scaler.fit_transform(df[["Attendance"]])
```

```python
pt = PowerTransformer()  #PowerTransformer is used to reduce skewness and make data closer to a normal distribution.
df[["Maths"]] = pt.fit_transform(df[["Maths"]])
```

```python
print("\nDataset After Transformation:")
print(df)
```
