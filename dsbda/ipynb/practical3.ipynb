```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("iris.csv")

print("First 5 Rows:")
print(df.head())

print("\nShape of Dataset:")
print(df.shape)

print("\nColumns:")
print(df.columns)
```

```python
print("Summary Statistics of Sepal Length by Species")
```

```python
group_stats = df.groupby("species")["sepal_length"].agg(
    ["mean", "median", "min", "max", "std"]
)

print(group_stats)
```

```python
print("\nNumeric Values of Sepal Length for Each Species:")

for sp in df["species"].unique():
    values = df[df["species"] == sp]["sepal_length"].tolist()
    
    print(f"\n{sp}:")
    print(values)
```

```python
species_list = ["setosa", "versicolor", "virginica"]

for sp in species_list:
    print("\n=====================================")
    print(f"Statistical Details of Iris-{sp}")
    print("=====================================")

    data = df[df["species"] == sp]

    print(data.describe(percentiles=[0.25, 0.50, 0.75]))
```

```python
print("\nMean of all Features Grouped by Species:")
print(df.groupby("species").mean())
```

```python
print("\nStandard Deviation of all Features Grouped by Species:")
print(df.groupby("species").std())
```
