```python
import nltk
```

```python
import pandas as pd

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('wordnet')
```

```python
text = "Ravi is reading a book. He likes reading books."

print("Original Text:\n")
print(text)
```

```python

tokens = word_tokenize(text)

print("\nTokens:")
print(tokens)
```

```python
pos_tags = pos_tag(tokens)

print("\nPOS Tags:")
print(pos_tags)
```

```python
stop_words = set(stopwords.words('english'))

filtered_words = [word for word in tokens if word.lower() not in stop_words and word.isalpha()]

print("\nAfter Stopwords Removal:")
print(filtered_words)
```

```python
ps = PorterStemmer()

stemmed_words = [ps.stem(word) for word in filtered_words]

print("\nStemmed Words:")
print(stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

print("\nLemmatized Words:")
print(lemmatized_words)
```

```python
documents = [
    "cat likes milk",
    "dog likes food",
    "cat and dog are animals"
]

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(documents)

# THIS LINE WAS MISSING
tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=vectorizer.get_feature_names_out()
)

feature_names = vectorizer.get_feature_names_out()

print("\nTF-IDF Feature Names:")
print(feature_names)

print("\nTF-IDF Matrix:")
print(tfidf_df)
```
